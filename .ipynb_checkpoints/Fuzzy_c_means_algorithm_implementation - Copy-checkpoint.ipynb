# **Calculating Cluster Center**

### To calculate centroids for each cluster we apply the following formula:

## C<sub>j</sub> = $ \frac{\sum_{i=1}^{n} W_{ij}^m x_{i}}{\sum_{i=1}^{n} W_{ij}^m} $

### m(fuzzy-ness) ranges from 1 to $\infty$

```python
def computeCentroids(weight_arr, data):
    C = []
    for i in range(k):
        weight_sum = np.power(weight_arr[:,i],m).sum()
        Cj = []
        for x in range(d):
            numerator = ( data[:,x] * np.power(weight_arr[:,i],m)).sum()
            c_val = numerator/weight_sum;
            Cj.append(c_val)
        C.append(Cj)
    return C 
```

```python
def updateCentroids(weight_arr,data):
    Wij_power_m = np.power(weight_arr,m) 
    term1 = data.T.dot(Wij_power_m).T
    term2 = Wij_power_m.sum(axis = 0)
    V = (term1.T / term2).T  
    return V     
```

```python
c1 = computeCentroids(weight_arr, df.iloc[:].values)
c2 = updateCentroids(weight_arr, df.iloc[:].values)

print(np.asarray(c1))
print(c2)
```

# **Updating Membership Value**

### Calculate the fuzzy-pseudo partition with the above formula
## $$w_{ij} = \frac{(\frac{1}{dist(x_i, c_j)})^{\frac{1}{m-1}}}{\sum_{s=1}^{k}(\frac{1}{dist(x_i,c_s)})^{\frac{1}{m-1}}}w 
ij$$


### where $dist (x_i, c_j)$ is the Euclidean distance between $x_{i}$ and $c_{j}$ cluster center.

```python
def updateWeights(weight_arr,C, data):
    denom = np.zeros(n)
    for i in range(k):
        dist = (data - C[i])**2
        dist = np.sum(dist, axis=1)
        dist = np.sqrt(dist)
        denom  = denom + np.power(1/dist,1/(m-1))

    for i in range(k):
        dist = (data - C[i])**2
        dist = np.sum(dist, axis=1)
        dist = np.sqrt(dist)
        weight_arr[:,i] = np.divide(np.power(1/dist,1/(m-1)),denom)
    return weight_arr
```

```python
def updateDistances(V,A, data):
    
    clusters = V.shape[0]
    samples = data.shape[0]
        
    dist = np.zeros((samples, clusters))
    
    for i in range(0, clusters):
        for j in range(0,samples):
            dist_L1 = data[j,:] - V[i]
            #print(df.iloc[j,:].values)
            dist_L2 = dist_L1.T.dot(dist_L1.dot(A))
            dist[j,i] = np.sqrt(dist_L2)

    return dist
        

def updateWeightsv2(weight_arr,V,dist, data):
    
    clusters = V.shape[0]
    samples = data.shape[0]
    
    U = np.zeros((weight_arr.shape[0], weight_arr.shape[1]))

    for i in range(0, clusters):
        for j in range(0,samples):
            term = 0
            for k in range(0, clusters):
                term += np.power(np.divide(dist[j,i],dist[j,k]),2/(m-1))
            U[j,i] = np.divide(1, term)
            
    return U    
```

# **Fuzzy algorithm**

```python
def FuzzyMeansAlgorithm(data):
    weight_arr = initializeMembershipWeights()
    plt.figure(figsize=(50,50)) 
    for z in range(MAX_ITERS):
        C = updateCentroids(weight_arr, data)
        A = np.identity(4, dtype = None)
        #u = updateWeights(weight_arr,C, df.iloc[:].values)
        dist = updateDistances(C,A, data)
        u = updateWeightsv2(weight_arr,C,dist, data)
        weight_arr = u
        plotData(z,C)
    plt.show()
    return (weight_arr,C)

```

# **Running algorithm**

```python
final_weights,Centers = FuzzyMeansAlgorithm(df.iloc[:].values)
```

```python
def vectorize(image):
    
    data = np.zeros((image.shape[0] * image.shape[1],5))
    
    for i in range(0, image.shape[0]):
        for j in range(0, image.shape[1]):
            data[i * image.shape[1] + j] = np.array([image[i][j][0],image[i][j][1],image[i][j][2],i,j])
    
    return data
```

```python
def argmax(arr):
    l = []
    for i in range(0, arr.shape[0]):
        l.append(list(np.where(arr[i] == np.amax(arr[i]))[0])[0])
    return l
```

```python
def generate_color(centeroid_number):
    colors = np.zeros((centeroid_number, 3))
    for i in range(0, centeroid_number):
        colors[i] = np.array([random.randint(0, 255) for _ in range(3)])
    return colors
```

```python
def devectorize(colors, pixel_segments, image):
    
    w = image.shape[1]
    h = image.shape[0]
    
    segmented_image = np.zeros((h, w, 3))
    for i in range(0, h):
        segmented_image[i] = np.array([colors[p] for p in pixel_segments[i * w: i * w + w]])
    return segmented_image
```
